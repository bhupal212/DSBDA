```python
import seaborn as sns
```

```python
dataset = sns.load_dataset("iris")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
import matplotlib.pyplot as plt
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,9))
sns.histplot(dataset['sepal_length'],ax=axes[0,0])
sns.histplot(dataset['sepal_width'],ax=axes[1,0])
sns.histplot(dataset['petal_length'],ax=axes[0,1])
sns.histplot(dataset['petal_length'],ax=axes[1,1])
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,9))
sns.boxplot(dataset['petal_length'],ax=axes[0,0])
sns.boxplot(dataset['sepal_width'],ax=axes[1,0])
sns.boxplot(dataset['petal_length'],ax=axes[0,1])
sns.boxplot(dataset['sepal_length'],ax=axes[1,1])
```
