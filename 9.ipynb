```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.describe()
```

```python
sns.boxplot(x='sex',y='age',data=titanic)
```

```python
sns.histplot(titanic['fare'],bins=20,kde=True)
plt.xlabel('fare')
plt.ylabel('freq')
```

```python
sns.scatterplot(x='class',y='fare',data=titanic)
plt.show()
```

```python
sns.scatterplot(x='sex',y='fare',data=titanic)
```
